# Triển khai hệ mã Diffie-Hellman

Đầu tiên Alice và Bob sẽ thống nhất chung một số nguyên tố p và 1 căn nguyên thủy g

```python
p = 23
g = 5
```

Alice chọn một số nguyên bí mật a = 6, và gửi cho Bob giá trị A = g^a mod p

```python
A = 5 ** 6 % 23

print(A)
```

Bob chọn một số nguyên bí mật b = 15, và gửi cho Alice giá trị B = g^b mod p

```python
B = 5 ** 15 % 23

print(B)
```

Alice sau khi nhận được B từ Bob thì tính s = B^a mod p

```python
s = 19 ** 6 % 23

s
```

Bob sau khi nhận được A từ Alice thì cũng tính s = A^b mod p

```python
s = 8 ** 15 % 23

s
```

Vậy cả hai đều có s = 2 rồi !!!
